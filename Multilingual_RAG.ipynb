```python
def doc_chunk(docs, chunk_size= 100, chunk_overlap= 50):
  # Split
  text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
      chunk_size = chunk_size,
      chunk_overlap = chunk_overlap)

  # Make splits
  splits = text_splitter.split_documents(docs)
  return splits
```

```python
french_splits = doc_chunk(french_docs)
english_splits = doc_chunk(english_docs)
german_splits = doc_chunk(german_docs)
```

```python
EMBEDDING_MODEL_NAME = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"

embeddings = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL_NAME
)
```

```python
pc = Pinecone(api_key=pinecone_api_key)

def create_pinecone_index(pc, index_name, embedding_dim = 1024, use_serverless = True, metric = "dotproduct"):
    if use_serverless:
        spec = ServerlessSpec(cloud='aws', region='us-west-2')
    else:
        # if not using a starter index, you should specify a pod_type too
        spec = PodSpec()
    # check for and delete index if already exists
    if index_name in pc.list_indexes().names():
        pc.delete_index(index_name)
    # create a new index
    pc.create_index(
        index_name,
        dimension= embedding_dim,  # dimensionality of the embedding model
        metric= metric,
        spec=spec
    )
    # wait for index to be initialized
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)
```

```python
create_pinecone_index(pc, "french-db", embedding_dim=384)
create_pinecone_index(pc, "english-db", embedding_dim=384)
create_pinecone_index(pc, "german-db", embedding_dim=384)
```

```python
from langchain.vectorstores import Pinecone
def create_pinecone_vectorstore(splits, embeddings, index_name):
  vectorstore = Pinecone.from_documents(splits, embeddings, index_name = index_name)
  retriever = vectorstore.as_retriever()
  return vectorstore, retriever
```

```python
french_vectorstore, french_retriever = create_pinecone_vectorstore(french_splits, embeddings, "french-db")
english_vectorstore, english_retriever = create_pinecone_vectorstore(english_splits, embeddings, "english-db")
german_vectorstore, german_retriever = create_pinecone_vectorstore(german_splits, embeddings, "german-db")
```

```python
routing_local_llm = "stablelm2"

routing_llm = ChatOllama(model=routing_local_llm, temperature=0.75)
```

```python
prompt = """
Answer in one line
what is the origin language of the following query:
{question}

The possible languages are:
- French
- English
- German
"""

prompt_template = PromptTemplate(
    input_variables=["question"],
    template=prompt,
)

router = prompt_template | routing_llm | StrOutputParser()

question = """
When was the human rights treaty made
"""

result = router.invoke({'question': question})
print(result)
```

```python
def hyde_doc_generation(llm, hyde_template):
  # HyDE document genration
  # template = """Please write a scientific paper passage to answer the question
  # Question: {question}
  # Passage:"""
  prompt_hyde = ChatPromptTemplate.from_template(hyde_template)

  generate_docs_for_retrieval = (
      prompt_hyde | llm | StrOutputParser()
  )

  return generate_docs_for_retrieval

def hyde_docs(generate_docs_for_retrieval, retriever):
  retrieval_chain = generate_docs_for_retrieval | retriever
  retrieved_docs = retrieval_chain.invoke({"question":question})
  return retrieved_docs

def hyde_rag(llm, rag_template):
  # template = """Answer the following question based on this context:

  # {context}

  # Question: {question}
  # """

  prompt = rag_template

  final_rag_chain = (
      prompt
      | llm
      | StrOutputParser()
  )

  return final_rag_chain

def rag_wrapper(llm, retriever, hyde_template, rag_template):
  generate_docs_for_retrieval = hyde_doc_generation(llm, hyde_template)
  retrieved_docs = hyde_docs(generate_docs_for_retrieval, retriever)
  rag_chain = hyde_rag(llm, rag_template)
  return rag_chain, retrieved_docs
```

```python
def call_hugging_face_pipeline(model_id):
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForCausalLM.from_pretrained(model_id)
    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=10)
    hf = HuggingFacePipeline(pipeline=pipe)
    return hf
```

```python
def choose_route(result):
    if "French" in result:
        french_hyde_template = """Veuillez rédiger un paragraphe scientifique pour répondre à la question
        Question: {question}
        Paragraphe:"""
        french_rag_template = """Répondez à la question suivante en fonction de ce contexte:

        {context}

        Question: {question}
        """
        french_rag_template = ChatPromptTemplate.from_template(french_rag_template)
        french_model_id = "paulml/Hermes-2-Pro-French"
        french_hyde_llm = call_hugging_face_pipeline(french_model_id)
        french_chain, retrieved_docs = rag_wrapper(french_hyde_llm,
                                                    french_retriever,
                                                    french_hyde_template,
                                                    french_rag_template)
        return french_chain, retrieved_docs
    elif "English" in result:
        hyde_local_llm = "mistral-openorca"

        english_hyde_llm = ChatOllama(model=hyde_local_llm, temperature=0.75)
        english_hyde_template = """Please write a scientific paper passage to answer the question
        Question: {question}
        Passage:"""
        english_rag_template = hub.pull("rlm/rag-prompt")
        english_chain, retrieved_docs = rag_wrapper(english_hyde_llm,
                                                    english_retriever,
                                                    english_hyde_template,
                                                    english_rag_template)
        return english_chain, retrieved_docs
    else:
        german_model_id = "mayflowergmbh/Wiedervereinigung-7b-dpo"
        german_hyde_llm = call_hugging_face_pipeline(german_model_id)
        german_hyde_template = """Bitte verfassen Sie eine wissenschaftliche Textpassage zur Beantwortung der Frage
        Frage: {question}
        Textpassage:"""
        german_rag_template = hub.pull("askmewhy/rag-prompt-de")
        german_chain, retrieved_docs = rag_wrapper(german_hyde_llm,
                                                    german_retriever,
                                                    german_hyde_template,
                                                    german_rag_template)
        return german_chain, retrieved_docs
```

```python
final_chain, retrieved_docs = choose_route(result)
```

```python
answer = final_chain.invoke({"context":retrieved_docs,"question":question})
```
